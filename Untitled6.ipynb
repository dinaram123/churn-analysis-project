```python
tele_encoded = tele.copy()

# Convert binary categorical columns to 0/1
binary_cols = ['Churn', 'Partner', 'Dependents', 'PhoneService',
               'PaperlessBilling', 'SeniorCitizen']

# Convert 'Yes'/'No' to 1/0
for col in binary_cols:
    if tele_encoded[col].dtype == 'object':
        tele_encoded[col] = tele_encoded[col].map({'Yes': 1, 'No': 0})

# Also encode 'Churn' for correlation (if still object)
if tele_encoded['Churn'].dtype == 'object':
    tele_encoded['Churn'] = tele_encoded[col].map({'Yes': 1, 'No': 0})

# Drop non-numeric columns before calculating correlation
tele_encoded = tele_encoded.drop('customerID', axis=1)

# Show correlation with churn
churn_corr = tele_encoded.corr()['Churn'].sort_values(ascending=False)
print(churn_corr)
```
